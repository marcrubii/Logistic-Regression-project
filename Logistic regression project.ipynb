To put us in context, let's create a logistic regression model where the model adds a decision boundary to our data. Thanks to this, we can predict whether a buyer ultimately ends up buying on our website, depending on the number of minutes they have been on the site and the number of items they have viewed.

**x_train** will be the minutes that our buyer spends on our website and the number of products you have viewed on the website.

**y_train** It will be the prediction of whether the buyer has purchased from the online store. **y = 0 --> has not bought**, **y = 1 --> has bought**

## 1. Import packages

```python
## Firt of all, we import our packages for our model to work correctly
import numpy as np
import matplotlib.pyplot as plt
from utils import * ## utils.py is used to load the data
import math
import copy
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

## 2. Plot the data

## 3. Train the model using Sklearn

```python
X_train, y_train = load_data("data/data.txt")
plot_data(X_train, y_train[:], pos_label = "Has bought", neg_label = "Has not bought")
plt.ylabel('number of articles viewed')
plt.xlabel('minutes on the website')
plt.legend(loc="upper right")
plt.show
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train) # With .fit we train our data until converged
```

```python
# You are getting the optimal values of the parameters w and b after training the model
w = lr_model.coef_.flatten()
b = lr_model.intercept_
print("The optimal value for w", w)
print("The optimal value for b", b)
```

```python
print(f"Number of iterations until it has converged: {lr_model.n_iter_}") # With .n_iter_ we can see the number of iterations until it has converged
```

```python
plot_decision_boundary(w, b, X_train, y_train)
# Set the y-axis label
plt.ylabel('number of articles viewed') 
# Set the x-axis label
plt.xlabel('minutes on the website') 
plt.legend(loc="upper right")
plt.ylim(ymin=0, ymax=15)  # We adjust the data to make it look better
plt.show()
```

## 4. Predictions

With Sklearn's **.predict** function, we can make our linear regression model "guess" the classes based on what it has learned during training. Remember, the first element are the number of articles viewed and the second are the minutes on the website

```python
y1_pred = lr_model.predict([[8, 2]]) # 8 articles and 2 minutes in the website
y2_pred = lr_model.predict([[3, 15]]) # 3 articles and 15 minutes in the website
print("The model for y1_pred has predicted", y1_pred, "the customer is not going to buy in the store")
print("The model for y1_pred has predicted", y2_pred, "the customer is going to buy in the store")
```

## 5 .Calculate accuracy

With the function **.score** we can see the accuracy of our training set

```python
print("Accuracy on training set:", lr_model.score(X_train, y_train))
```
